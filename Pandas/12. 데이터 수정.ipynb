# 12. 데이터 수정

```python
import pandas as pd
df = pd.read_excel('score.xlsx', index_col='지원번호')
df
```

## Column 수정

```python
df['학교'].replace({'북산고':'상북고'}) # 북산고 -> 상북고로 replace
```

```python
df['학교'].replace({'북산고':'상북고', '능남고':'남강고'})
```

```python
df['학교'].replace({'북산고':'상북고'}, inplace=True)
```

```python
df
```

```python
df['SW특기'].str.lower()
```

```python
df['SW특기'] = df['SW특기'].str.upper()
```

```python
df
```

```python
df['학교'] = df['학교'] + '등학교' # 학교 데이터 + 등학교
df
```

## Column 추가

```python
df['총합'] = df['국어'] + df['영어'] + df['수학'] + df['과학'] + df['사회']
df
```

```python
df['결과'] = 'Fail' # 결과 Column 추가 후 Fail 입력
df
```

```python
df.loc[df['총합'] > 400, '결과'] = 'Pass' # 총합이 400 이상이면 결과 Column 에 'Pass' 입력
df
```

## Column 삭제

```python
df.drop(columns=['총합']) # 총합 Column 삭제
```

```python
df.drop(columns=['국어', '영어', '수학']) # 국어, 영어, 수학 삭제 <- 이것도 'inplace=True' 필요
```

## Row 삭제

```python
df.drop(index='4번') # 4번 학생 데이터 row를 삭제
```

```python
filt = df['수학'] < 80 # 수학 점수 80 점 미만 학생 필터링
df[filt]
```

```python
df[filt].index
```

```python
df.drop(index=df[filt].index)
```

## Row 추가

```python
df.loc['9번'] = ['이정환', '해남고등학교', 184, 90, 90, 90, 90, 90, 'Kotlin', 450, 'Pass'] # 새로운 Row 추가
df
```

## Cell 수정

```python
df.loc['4번', 'SW특기'] = 'Python' # 4번 학생의 SW특기를 Python으로 변경
df
```

```python
df.loc['5번', ['학교', 'SW특기']] = ['능남고등학교', 'C'] # 5번 학생 학교 = 능남, 특기 = C 수정
df
```

## Column 순서변경

```python
cols = list(df.columns)
cols
```

```python
df = df[[cols[-1]] + cols[0:-1]] # 맨 뒤에 있는 결과 Column 맨 앞으로 가져오기 [cols[-1]] 의 경우 리스트로 만들어주기 위해 대괄호 입력
df
```
