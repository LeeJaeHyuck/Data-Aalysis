# 4. 파일 저장 및 열기
DataFrame 객체를 Excel, csv, txt 등 형태의 파일로 저장 및 열기

```python
import pandas as pd

data = {
    '이름' : ['채치수', '정대만', '송태섭', '서태웅', '강백호', '변덕규', '황태산', '윤대협'],
    '학교' : ['북산고', '북산고', '북산고', '북산고', '북산고', '능남고', '능남고', '능남고'],
    '키' : [197, 184, 168, 187, 188, 202, 188, 190],
    '국어' : [90, 40, 80, 40, 15, 80, 55, 100],
    '영어' : [85, 35, 75, 60, 20, 100, 65, 85],
    '수학' : [100, 50, 70, 70, 10, 95, 45, 90],
    '과학' : [95, 55, 80, 75, 35, 85, 40, 95],
    '사회' : [85, 25, 75, 80, 10, 80, 35, 95],
    'SW특기' : ['Python', 'Java', 'Javascript', '', '', 'C', 'PYTHON', 'C#']
}

df = pd.DataFrame(data, index=['1번', '2번', '3번', '4번', '5번', '6번', '7번', '8번'])
df.index.name = '지원번호'
df
```

## 저장하기
### csv 파일로 저장

```python
df.to_csv('score.csv', encoding='utf-8-sig')
```

```python
df.to_csv('score.csv', encoding='utf-8-sig', index=False)
```

### 텍스트(.txt) 파일로 저장

```python
df.to_csv('score.txt', sep='\t') # Tab으로 구분하여 텍스트 파일로 저장
```

### 엑셀 파일로 저장

```python
df.to_excel('score.xlsx')
```

***
## 열기

### CSV 파일 열기

```python
df = pd.read_csv('score.csv')
df
```

```python
df = pd.read_csv('score.csv', skiprows=1) # 지정된 개수 만큼의 Row를 건너뜀 (헤더가 첫번째라 헤더가 날라감)
# skiprows 의 경우 숫자 하나만 넣는 경우 헤더까지 카운트가 되고 [...] 으로 입력할 경우 헤더가 자동으로 인식 됨
```

```python
# [1,3,5] row 제외 (이렇게 수행할 경우 헤더는 자동인식 되어 데이터 row 1부터 시작 됨)
df = pd.read_csv('score.csv', skiprows=[1, 3, 5]) 
```

```python
df = pd.read_csv('score.csv', nrows=4) # 지정된 개수 만큼의 row 만 가져옴
```

```python
df = pd.read_csv('score.csv', skiprows=2, nrows=4) # 처음 두개 무시 후 4개만 가져오기
df
```

### txt 파일 열기

```python
df = pd.read_csv('score.txt', sep='\t')
df
```

```python
df = pd.read_csv('score.txt', sep='\t', index_col='지원번호') # index 지원번호로 입력
df
```

```python
df = pd.read_csv('score.txt', sep='\t')
df.set_index('지원번호', inplace=True)
df
```

### 엑셀 파일 열기

```python
df = pd.read_excel('score.xlsx')
df
```

```python
df = pd.read_excel('score.xlsx', index_col='지원번호')
df
```
