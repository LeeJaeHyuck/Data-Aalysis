## 데이터 채우기 fillna

```python
df.fillna('') # NaN 데이터를 빈 칸으로 채움
```

```python
df.fillna('없음')
```

```python
import numpy as np
df['학교'] = np.nan # 학교 데이터 전체를 NaN 으로 채움
df
```

```python
df.fillna('모름')
```

```python
df.fillna('모름', inplace=True) # inplace 해야 데이터 반영 됨
```

```python
df
```

```python
df['SW특기'].fillna('확인 중', inplace=True)  # SW특기 데이터 중에서 NaN 에 대해서 채움
```

## 데이터 제외하기 dropna

```python
df = pd.read_excel('score.xlsx', index_col='지원번호')
df
```

```python
df.dropna(inplace=True) # NaN 을 포함하는 데이터를 삭제
```

```python
df
```

```python
df = pd.read_excel('score.xlsx', index_col='지원번호')
```

- axis : index or columns
- how : any or all (any 가 default임)

```python
df.dropna(axis='index', how='any') # NaN 이 하나라도 있는 row 삭제
```

```python
df.dropna(axis='columns') # NaN 이 하나라도 있는 column 삭제
```

```python
df['학교'] = np.nan
df
```

```python
df.dropna(axis='columns', how='all') # 데이터 전체가 NaN 일 경우 column 자체 삭제
```
