## 데이터에 함수 적용 (apply)

```python
# Height 에 단위(cm) 붙이기
def add_cm(height):
    return str(height) + 'cm'

df['Height'] = df['Height'].apply(add_cm) # 키 데이터에 대해서 add_cm 함수를 호출한 결과 데이터를 반영
df
```

```python
# 첫 글자는 대문자로, 나머지는 소문자로 변경하는 함수
def capitalize(lang):
    if pd.notnull(lang): # NaN 확인
        return lang.capitalize() # Stirng 내장 함수임 내가 선언한거랑 다름
    return lang

df['SoftWare'] = df['SoftWare'].apply(capitalize)
df
```

```python
df['SoftWare'] = df['SoftWare'].str.capitalize()
```
