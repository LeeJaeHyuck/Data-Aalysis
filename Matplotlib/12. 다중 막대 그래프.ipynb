```python
index = np.arange(df.shape[0])
index
```

```python
w = 0.25
plt.bar(index - w, df['국어'])
plt.bar(index, df['영어'])
plt.bar(index + w, df['수학'])
```

```python
w = 0.25
plt.bar(index - w, df['국어'], width=w, label='국어')
plt.bar(index, df['영어'], width=w, label='영어')
plt.bar(index + w, df['수학'], width=w, label='수학')
plt.legend()
```

```python
plt.figure(figsize=(10, 5))
plt.title('학생별 성적')

w = 0.25
plt.bar(index - w, df['국어'], width=w, label='국어')
plt.bar(index, df['영어'], width=w, label='영어')
plt.bar(index + w, df['수학'], width=w, label='수학')
plt.legend(ncol=3)
plt.xticks(index, df['이름'], rotation=45)
plt.show()
```

```python
plt.figure(figsize=(5, 10))
plt.title('학생별 성적')

w = 0.25
plt.barh(index - w, df['국어'], height=w, label='국어')
plt.barh(index, df['영어'], height=w, label='영어')
plt.barh(index + w, df['수학'], height=w, label='수학')
plt.legend(ncol=3)
plt.yticks(index, df['이름'])
plt.show()
```
