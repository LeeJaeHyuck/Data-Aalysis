# 16. 여러 그래프

```python
import matplotlib.pyplot as plt
import matplotlib
import pandas as pd
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # Windows 한글 폰트
matplotlib.rcParams['font.size'] = 15 # 글자 크기
matplotlib.rcParams['axes.unicode_minus'] = False # 한글 폰트 사용 시, 마이너스 글자 깨지는 현상 방지

df = pd.read_excel('../Pandas/score.xlsx')
df['학년'] = [3, 2, 2, 1, 1, 2, 2, 1]
df
```

```python
fig, axs = plt.subplots(2, 2, figsize=(15, 13)) # 2 x 2 에 해당하는 plot 들을 생성
fig.suptitle('여러 그래프 넣기')

# 첫 번째 그래프 (세로 막대 그래프)
axs[0, 0].bar(df['이름'], df['국어'], label='국어점수') # 데이터 설정
axs[0, 0].set_title('학생별 국어 점수') # 제목
axs[0, 0].legend() # 범례
axs[0, 0].set(xlabel='이름', ylabel='점수') # x, y 축 label
axs[0, 0].set_facecolor('lightyellow') # 전면 색
axs[0, 0].grid(ls='--', lw=0.5, alpha=0.3)

# 두 번째 그래프 (꺾은 선 그래프)
axs[0, 1].plot(df['이름'], df['수학'], label='수학', marker='o')
axs[0, 1].plot(df['이름'], df['영어'], label='영어', marker='v')
axs[0, 1].legend()
axs[0, 1].set_title('학생별 수학, 영어 점수')
axs[0, 1].grid(ls='--', lw=0.5, alpha=0.7)

# 세 번째 그래프 (가로 막대 그래프)
axs[1, 0].barh(df['이름'], df['키'], height=0.75)
axs[1, 0].set_title('학생별 키')
axs[1, 0].grid(ls='--', lw=0.5, alpha=0.7)

# 네 번째 그래프
from collections import Counter
cnt = Counter(df['학년'])
values = [cnt[1], cnt[2], cnt[3]]
labels = ['1학년', '2학년', '3학년']
explode = [0.05] * len(values)
axs[1, 1].pie(values, labels=labels, autopct='%.1f%%', explode=explode)
axs[1, 1].set_title('학년별 학생 분포')
```
