# 4. 스타일

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # Windows 한글 폰트
matplotlib.rcParams['font.size'] = 15 # 글자 크기
matplotlib.rcParams['axes.unicode_minus'] = False # 한글 폰트 사용 시, 마이너스 글자 깨지는 현상 방지

x = [1, 2, 3]
y = [2, 4, 8]
```

```python
plt.plot(x, y)
```

```python
plt.plot(x, y, linewidth=5) # 라인 두께
```

## Marker

```python
plt.plot(x, y, marker='o') # 그래프에 동그라미 표시
```

```python
plt.plot(x, y, marker='o', linestyle='None') # 라인 없이 점만 표시
```

```python
plt.plot(x, y, marker='v') # 삼각형
```

```python
plt.plot(x, y, marker='v', markersize=10)
```

```python
plt.plot(x, y, marker='X', markersize=10)
```

```python
plt.plot(x, y, marker='o', markersize=10, markeredgecolor='red') # marker 테두리 색
```

```python
plt.plot(x, y, marker='o', markersize=10, markeredgecolor='red', markerfacecolor='yellow') # maker 색
```

## 선 스타일

```python
plt.plot(x, y, linestyle=':')
```

```python
plt.plot(x, y, linestyle='--')
```

```python
plt.plot(x, y, linestyle='-.')
```

## 색깔

```python
plt.plot(x, y, color='red')
```

```python
plt.plot(x, y, color='#f00')
```

```python
plt.plot(x, y, color='b')
```

```python
plt.plot(x, y, color='g')
```

## 포맷

```python
plt.plot(x, y, 'ro--') # 'color, marker, linestyle'
```

```python
plt.plot(x, y, 'bv:')
```

```python
plt.plot(x, y, 'go') # Linestyle 없어서 선 안나옴
```

## 축약어

```python
# mfc = markerfacecolor, ms = markersize, mec = markeredgecolor, ls = linestyle
plt.plot(x, y, marker='o', mfc='red', ms=10, mec='y', ls=':')
```

## 투명도

```python
plt.plot(x, y, marker='o', mfc='r', ms=10, alpha=0.3) # alpha : 투명도 (0~1)
```

## 그래프 크기

```python
plt.figure(figsize=(10, 5))
plt.plot(x, y)
```

```python
plt.figure(figsize=(5, 10))
plt.plot(x, y)
```

```python
plt.figure(figsize=(5, 5), dpi=50) # dpi : dots per inch, 확대
plt.plot(x, y)
```

## 배경색

```python
plt.figure(facecolor='skyblue')
plt.plot(x, y)
```
