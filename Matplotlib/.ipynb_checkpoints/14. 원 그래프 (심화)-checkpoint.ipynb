# 14. 원 그래프 (심화)

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # Windows 한글 폰트
matplotlib.rcParams['font.size'] = 13 # 글자 크기
matplotlib.rcParams['axes.unicode_minus'] = False # 한글 폰트 사용 시, 마이너스 글자 깨지는 현상 방지
```

```python
values = [30, 25, 20, 13, 10, 2]
labels = ['Python', 'Java', 'Javascript', 'C#', 'C/C++', 'ETC']
colors = ['#ffadad', '#ffd6a5', '#fdffb6', '#caffbf', '#9bf6ff', '#a0c4ff']
explode = [0.05] * len(values)

# autopct : 숫자 넣기, startangle : 기울기, counterclock : 방향
plt.pie(values, labels=labels, autopct='%.1f%%', startangle=90, counterclock=False, colors=colors, explode = explode)
plt.show()
```

```python
wedgeprops={'width':0.6}
plt.pie(
    values,
    labels=labels,
    autopct='%.1f%%',
    startangle=90,
    counterclock=False,
    colors=colors,
    wedgeprops=wedgeprops # 가운데 구멍내기
)
plt.show()
```

```python
wedgeprops={'width':0.6, 'edgecolor':'w', 'linewidth': 3}
plt.pie(
    values,
    labels=labels,
    autopct='%.1f%%',
    startangle=90,
    counterclock=False,
    colors=colors,
    wedgeprops=wedgeprops
)
plt.show()
```

```python
def custom_autopct(pct):
    return ('%.1f%%' % pct) if pct >= 10 else ''

plt.pie(
    values,
    labels = labels,
    autopct = custom_autopct,
    startangle = 90,
    counterclock = False,
    colors = colors,
    wedgeprops = wedgeprops,
    pctdistance = 0.7 # 숫자 중심으로부터 떨어트리기
)
plt.show()
```

## DataFrame 활용

```python
import pandas as pd
df = pd.read_excel('../Pandas/score.xlsx')
df
```

```python
grp = df.groupby('학교')
grp
```

```python
grp.size()['북산고']
```

```python
values = [grp.size()['북산고'], grp.size()['능남고']]
labels = ['북산고', '능남고']

plt.pie(values,
    labels = labels,
    autopct = '%.1f%%',
    startangle = 90,
    counterclock = False,
    colors = ['c', 'y'],
    wedgeprops = {'width':0.6, 'edgecolor':'r', 'linewidth': 3},
    pctdistance = 0.7
       )
plt.title('소속 학교')
plt.show()
```
