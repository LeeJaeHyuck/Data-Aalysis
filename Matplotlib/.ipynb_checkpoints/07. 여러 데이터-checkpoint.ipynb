## COVID-19 백신 종류별 접종 인구

```python
days = [1, 2, 3]
az = [2, 4, 8] # 아스트라제네카
pfizer = [5, 1, 3] # 화이자
moderna = [1, 2, 5] # 모더나

plt.plot(days, az, label='아스트라제네카')
plt.plot(days, pfizer, label='화이자')
plt.plot(days, moderna, label='모더나')
plt.legend()
```

```python
plt.plot(days, az, label='아스트라제네카', marker='v')
plt.plot(days, pfizer, label='화이자', marker='o', ls='--')
plt.plot(days, moderna, label='모더나', marker='s', ls='-.')

plt.legend()
```

```python
plt.plot(days, az, label='아스트라제네카', marker='v')
plt.plot(days, pfizer, label='화이자', marker='o', ls='--')
plt.plot(days, moderna, label='모더나', marker='s', ls='-.')

plt.legend(ncol=3) # 범례 컬럼 3개까지 표시
```
