# 8. 막대 그래프(기본)

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # Windows 한글 폰트
matplotlib.rcParams['font.size'] = 15 # 글자 크기
matplotlib.rcParams['axes.unicode_minus'] = False # 한글 폰트 사용 시, 마이너스 글자 깨지는 현상 방지
```

```python
labels = ['강백호', '서태웅', '정대만'] # 이름
values = [190, 187, 184] # 키

plt.bar(labels, values)
```

```python
labels = ['Kang', 'Soe', 'Jung'] # 이름
values = [190, 187, 184] # 키

plt.bar(labels, values, color='r')
```

```python
labels = ['Kang', 'Soe', 'Jung'] # 이름
values = [190, 187, 184] # 키
colors = ['r', 'g', 'b']

plt.bar(labels, values, color=colors, alpha=0.7)
```

```python
labels = ['Kang', 'Soe', 'Jung'] # 이름
values = [190, 187, 184] # 키

plt.bar(labels, values)
plt.ylim(150, 195)
```

```python
plt.bar(labels, values, width=0.5)
```

```python
plt.bar(labels, values, width=0.3)
plt.xticks(rotation=45) # x 축의 이름 데이터 각도를 45도로 설정
plt.yticks(rotation=45) # y 축의 이름 데이터 각도를 45도로 설정
```

```python
ticks = ['1번학생', '2번학생', '3번학생']

plt.bar(labels, values)
plt.xticks(labels, ticks)
```
